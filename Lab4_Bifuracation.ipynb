```python
def F(x, r, p, a):
  return a*x * (1-x) * (2*x - 1) + r * (p - x)

def Froots(r, p, a):
  coeff = [-2*a, 3*a, -(a+r), r * p]
  return np.roots(coeff)

def delF(x, r, p, a):
  return -6*a*(x**2) + (6*a*x) - (a + r)
```

```python
a = 1
p = np.linspace(0, 1, 500)
r = 0.6

x = []
stablep = []
stablex = []
unstablep = []
unstablex = []
for i in range(len(p)):
  roots = Froots(r, p[i], a)
  rroots = np.array([])
  x_eq = []

  for j in range(len(roots)):
    if (np.isreal(roots[j])):
      if(delF(roots[j], r, p[i], a)>0):
        unstablex = np.append(roots[j], unstablex)
        unstablep = np.append(p[i], unstablep)
      elif(delF(roots[j], r, p[i], a)<=0):
        stablex = np.append(roots[j], stablex)
        stablep = np.append(p[i], stablep)
  plt.scatter(stablep, stablex, s=2, color='blue', marker='o')
  plt.scatter(unstablep, unstablex, s=2, color='red', marker='o')
  plt.legend(["stable", "unstable"])
  plt.grid(True)
  plt.xlabel('p')
  plt.ylabel('x_eq')
  plt.title('Bifurcation Diagram')
```
