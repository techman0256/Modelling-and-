```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def aplha(p , q, N, C):
  return p + (q * (N/C))
```

```python
def function(p, q, N_0, C, length):
  delT = 0.1
  N = N_0

  time = []
  users = []
  changes = []
  # print(C - N)
  for i in np.arange(0, length, delT):
    time = np.append(i, time)

    dN = aplha(p, q, N, C) * (C - N) * delT
    N = N + (dN)
    # print(N, C - N)
    users = np.append(N, users)
    changes = np.append(dN, changes)

  plt.plot(time, users, color='g')
#   plt.plot(time, changes, color='k')
  # plt.subplot(121)
  plt.xlabel('Time')
  plt.ylabel('Number of Users')
  plt.legend()
  plt.grid(True)
  plt.show()

  plt.plot(time, changes, color='k')
  # plt.axvline(x = C/2)
  plt.xlabel('Time')
  plt.ylabel('dN')
  plt.legend()
  plt.grid(True)
  plt.show()
```

External Influence Model for different p values

```python
function(0.0, 0.21566, 20, 40000000, 180)
```
